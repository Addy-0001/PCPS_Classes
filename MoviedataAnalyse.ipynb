```python
# Import necessary libraries
import pandas as pd

# Load the dataset from your Google Drive
file_path = './movies_dataset.csv'  # Update the path according to your file location
df = pd.read_csv(file_path)

# Display the first few rows of the dataset to ensure it's loaded correctly
print("First few rows of the dataset:")
print(df.head())
```

#Analysing data using pandas


```python
# Get basic information about the dataset
print("\nBasic Information:")
print(df.info())

# Get summary statistics of numerical columns
print("\nSummary Statistics:")
print(df.describe())

# Check for any missing values in the dataset
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Count the number of movies in each genre
genre_counts = df['Genre'].value_counts()

# Display the distribution of movies by genre
print("\nDistribution of Movies by Genre:")
print(genre_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plots
sns.set(style="whitegrid")
```

```python
# Plot the distribution of movies by genre
plt.figure(figsize=(8, 6))
sns.countplot(x='Genre', data=df, palette='viridis')
plt.title('Distribution of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the average IMDb rating by genre
avg_rating_by_genre = df.groupby('Genre')['IMDb Rating'].mean().sort_values()

# Plot the average IMDb rating by genre
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_rating_by_genre.index, y=avg_rating_by_genre.values, palette='coolwarm')
plt.title('Average IMDb Rating by Genre')
plt.xlabel('Genre')
plt.ylabel('Average IMDb Rating')
plt.ylim(5, 9)
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of movies released each year
movies_by_year = df['Release Year'].value_counts().sort_index()

# Plot the number of movies released over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x=movies_by_year.index, y=movies_by_year.values, marker='o', color='blue')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
# Calculate the average IMDb rating by director
avg_rating_by_director = df.groupby('Director')['IMDb Rating'].mean().sort_values()

# Plot the average IMDb rating by director
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_rating_by_director.index, y=avg_rating_by_director.values, palette='magma')
plt.title('Average IMDb Rating by Director')
plt.xlabel('Director')
plt.ylabel('Average IMDb Rating')
plt.ylim(5, 9)
plt.xticks(rotation=45)
plt.show()
```
